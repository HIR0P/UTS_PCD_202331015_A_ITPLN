```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca gambar
image_path = 'gambar2.jpg' 
image_backlight = cv2.imread(image_path)
```

```python
# Konversi gambar menjadi grayscale
image_gray = cv2.cvtColor(image_backlight, cv2.COLOR_BGR2GRAY)
```

```python
# Meningkatkan kecerahan gambar
image_bright = cv2.convertScaleAbs(image_gray, alpha=1.5, beta=50)
```

```python
# Meningkatkan kontras menggunakan histogram equalization
image_contrast = cv2.equalizeHist(image_bright)
```

```python
# Menampilkan gambar asli, grayscale, brightened, dan contrast-enhanced
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Tampilkan gambar asli
axes[0, 0].imshow(cv2.cvtColor(image_backlight, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title("Gambar Asli")
axes[0, 0].axis('off')

# Tampilkan gambar grayscale
axes[0, 1].imshow(image_gray, cmap='gray')
axes[0, 1].set_title("Gambar Gray")
axes[0, 1].axis('off')

# Tampilkan gambar yang dipercerah (brightened)
axes[1, 0].imshow(image_bright, cmap='gray')
axes[1, 0].set_title("Gambar Dipercerah")
axes[1, 0].axis('off')

# Tampilkan gambar yang diperkontras (contrast-enhanced)
axes[1, 1].imshow(image_contrast, cmap='gray')
axes[1, 1].set_title("Gambar dikontraskan")
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```
